# API Testing script

Script used to test the api and understand how to import the data as wanted

If possible use a conda environment as explained in the `setup_instruction.md`

```python
# imports
import pandas as pd
import numpy as np
import os
import volue_insight_timeseries
from dotenv import load_dotenv

load_dotenv()
# load environment variables
CLIENT_ID = os.getenv("VOLUE_CLIENT_ID")
CLIENT_SECRET = os.getenv("VOLUE_CLIENT_SECRET")
```

```python
def main():
    try:
        # 1. Initialize the session
        print("Authenticating...")
        session = volue_insight_timeseries.Session(
            client_id=CLIENT_ID,
            client_secret=CLIENT_SECRET
        )
        print("Authentication succeeded.")
 
        # 2. Define the curve name to download
        # NOTE: this is the example from your snippet.
        # If you get a "Curve not found" error, replace this name with one valid for your subscription.
        curve_name = 'pri ch spot ec00 €/mwh cet h f'
        
        #curve_name = 'pri ch spot ec00 €/mwh cet h a'
        
        print(f"Retrieving curve object: {curve_name}")
        curve = session.get_curve(name=curve_name)
        print(curve.id)
        
        # Print curve type (TIME_SERIES, TAGGED, etc.)
        print(f"Curve type: {curve.curve_type}")
    
        ts = curve.get_instance(issue_date=pd.Timestamp('2026-01-01T00:00'))
 
        ts = curve.search_instances(issue_date_from='2026-01-01', issue_date_to='2026-01-02', with_data=True)
        print(ts[0].to_pandas())
    except volue_insight_timeseries.exceptions.AuthenticationError as auth_err:
        print(f"Authentication error: {auth_err}")

main()
```
